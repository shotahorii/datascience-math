## 1. Definition 

Below 4 definitions are equivalent.

A $d \times d$ Matrix $M$ is an orthogonal matrix if...

### Definition 1

$M^T = M^{-1}$

In other words, $M^T M=MM^T = I$

### Definition 2

$d$ column vectors of $M$ are a set of orthonormal basis.

Let: ${\bf c_1},{\bf c_2},...,{\bf c_d}$ be column vectors of $M$.

$
M^TM = \begin{bmatrix}
{\bf c_1}^T{\bf c_1} & {\bf c_1}^T{\bf c_2} & \cdots & {\bf c_1}^T{\bf c_d} \\
{\bf c_2}^T{\bf c_1} & {\bf c_2}^T{\bf c_2} & \cdots & {\bf c_2}^T{\bf c_d} \\
\vdots & \vdots & \ddots & \vdots \\
{\bf c_d}^T{\bf c_1} & {\bf c_d}^T{\bf c_2} & \cdots & {\bf c_d}^T{\bf c_d} \\
\end{bmatrix} = I
$

Hence, ${\bf c_i}^T{\bf c_j} = \delta_{ij}$ where, $\delta_{ij}=1$ when $i=j$, and $\delta_{ij}=0$ when $i \ne j$

This means $d$ column vectors of $M$ are a set of orthonormal basis.

### Definition 3

$d$ row vectors of $M$ are a set of orthonormal basis.

Let: ${\bf r_1},{\bf r_2},...,{\bf r_d}$ be row vectors of $M$.

$
MM^T = \begin{bmatrix}
{\bf r_1}^T{\bf r_1} & {\bf r_1}^T{\bf r_2} & \cdots & {\bf r_1}^T{\bf r_d} \\
{\bf r_2}^T{\bf r_1} & {\bf r_2}^T{\bf r_2} & \cdots & {\bf r_2}^T{\bf r_d} \\
\vdots & \vdots & \ddots & \vdots \\
{\bf r_d}^T{\bf r_1} & {\bf r_d}^T{\bf r_2} & \cdots & {\bf r_d}^T{\bf r_d} \\
\end{bmatrix} = I
$

Hence, ${\bf r_i}^T{\bf r_j} = \delta_{ij}$ where, $\delta_{ij}=1$ when $i=j$, and $\delta_{ij}=0$ when $i \ne j$

This means $d$ row vectors of $M$ are a set of orthonormal basis.

### Definition 4

$M$ is a linear transformation that keeps the value of inner product of vetors.

Let: ${\bf v}, {\bf w} \in $ a vector space $V$, where inner product is $({\bf v}, {\bf w})={\bf v}^T{\bf w}$

$(M{\bf v}, M{\bf w}) = (M{\bf v})^TM{\bf w} = {\bf v}^TM^TM{\bf w} = {\bf v}^T{\bf w} = ({\bf v}, {\bf w})$

## 2. Properties 

### Property 1

$|M| = \pm 1$

**Proof**

Note: $|AB| = |A||B|$ and $|A^T|=|A|$

$1 = |I| = |M^TM|=|M^T||M| = |M|^2$ 

### Property 2

Let: ${\bf v}$ be a $d$-dimensional vector and $||\cdot||$ be norm.

$||M{\bf v}||=||{\bf v}||$

**Proof**

$||M{\bf v}|| = \sqrt{(M{\bf v})^TM{\bf v}} = \sqrt{{\bf v}^TM^TM{\bf v}} = \sqrt{{\bf v}^T{\bf v}}=||{\bf v}||$

### Property 3

$M^{-1}$ is also an orthogonal matrix.

**Proof**

(By definition 2) $M$'s column vectors are a set of orthogonal basis. 

$\Longleftrightarrow$ $M^T$'s row vectors are a set of orthogonal basis.

$\Longleftrightarrow$ (By definition 1) $M^{-1}$'s row vectors are a set of orthogonal basis.

$\Longleftrightarrow$ (By definition 3) $M^{-1}$ is an orthogonal matrix.
