## 1. Gamma function
### 1.1. Definition
For complex numbers with a positive real part $z$,

$ \Gamma(z) = \int^\infty_0 \mu^{z-1}e^{-\mu}d\mu$

$where \,\,\,\, Re(z) > 0 $

Also below equation is true. 

$ \Gamma(z+1) = z\Gamma(z) $

**Proof**

$ \Gamma(z+1) =  \int^\infty_0 \mu^z e^{-\mu}d\mu$

$ = \int^\infty_0 \mu^z (-e^{-\mu})'d\mu $

Using integration by parts,

$ = [-\mu^z e^{-\mu}]^\infty_0 + \int^\infty_0 z\mu^{z-1} e^{-\mu}d\mu $

$ = \lim\limits_{\mu \to \infty}(-\mu^ze^{-\mu}) - (-0e^{-0}) + \int^\infty_0 z\mu^{z-1} e^{-\mu}d\mu $

Note: $ \lim\limits_{x \to \infty}(-x^ze^{-x}) = 0 $

$ = \int^\infty_0 z\mu^{z-1} e^{-\mu}d\mu $

$ = z\int^\infty_0 \mu^{z-1} e^{-\mu}d\mu = z\Gamma(z)$

---

### 1.2. When $z$ is a natural number

$ \Gamma(n) = (n-1)! $

$ where\,\,\,\, n \in \mathbb{N} $

**Proof**

$ \Gamma(1) = \int^\infty_0 \mu^{1-1}e^{-\mu}d\mu = \int^\infty_0 e^{-\mu}d\mu$

$ = \lim\limits_{\mu \to \infty}(-e^{-\mu}) - (-e^{-0}) = 0 - (-1) = 1$

Given that $ \Gamma(1) = 1 $ and $ \Gamma(z+1) = z\Gamma(z) $,

$ \Gamma(n) = 1\cdot2\cdot3\cdot\cdot\cdot(n-1) = (n-1)!$

## 2. Beta function
### 2.1. Definition
For complex numbers with a positive real part $a,b$,

$ B(a,b) = \int^1_0 \mu^{a-1}(1-\mu)^{b-1}d\mu$

$where \,\,\,\, Re(a) > 0 \,\,\,\, Re(b) > 0$

---

### 2.2. Symmetricity

$B(a,b) = B(b,a)$

**Proof**

$B(a,b) = \int^1_0 \mu^{a-1}(1-\mu)^{b-1}d\mu$

Let: $ \mu' = 1-\mu$

$ = \int^1_0 (1-\mu')^{a-1}(\mu')^{b-1}d\mu' = B(b,a)$

## 3. Relationship between Gamma function and Beta function
### 3.1. Change of variable for Gamma function

$ \Gamma(z) = \int^\infty_0 \mu^{z-1}e^{-\mu}d\mu$

Change the variable $\mu$ as below.

$\mu = r^2 \,\,\, where \,\,\, r \ge 0$

Interval doesn't change. $\mu \in [0,\infty] \Longrightarrow r^2 \in [0,\infty]$

Note: $\int^b_a f(x)dx = \int^\beta_\alpha f(g(t))\frac{dx}{dt}dt = \int^\beta_\alpha f(g(t))g'(t)dt \,\,\, where \,\,\, g(\alpha)=a, g(\beta)=b$

$ \Gamma(z) = \int^\infty_0 (r^2)^{z-1}e^{-r^2}\frac{d}{dr}(r^2)dr 
= 2\int^\infty_0 r^{2z-2}e^{-r^2}rdr = 2\int^\infty_0 r^{2z-1}e^{-r^2}dr$

---

### 3.2. Change of variable for Beta function

$B(a,b) = \int^1_0 \mu^{a-1}(1-\mu)^{b-1}d\mu$

Change the variable $\mu$ as below.

$\mu = sin^2\theta$

Interval changes as $\mu \in [0,1] \Longrightarrow \theta \in [0,\frac{\pi}{2}]$

$B(a,b) = \int^{\frac{\pi}{2}}_0 (sin^2\theta)^{a-1}(1-sin^2\theta)^{b-1}\frac{d}{d\theta}(sin^2\theta)d\theta$

$=\int^{\frac{\pi}{2}}_0 (sin^2\theta)^{a-1}(cos^2\theta)^{b-1}2sin\theta cos\theta d\theta$

$=\int^{\frac{\pi}{2}}_0 sin^{2a-2}\theta \, cos^{2b-2}\theta \, 2sin\theta cos\theta d\theta$

$=2\int^{\frac{\pi}{2}}_0 sin^{2a-1}\theta \, cos^{2b-1}\theta \, d\theta$

---

### 3.3. Relationship equation

$\Gamma(a)\Gamma(b) = 2\int^\infty_0 u^{2a-1}e^{-u^2}du \cdot 2\int^\infty_0 v^{2b-1}e^{-v^2}dv$

$= 4\int^\infty_0\int^\infty_0 e^{-(u^2+v^2)} u^{2a-1} v^{2b-1} dudv$

Change the variables $u,v$ as below. (Polar conversion)

$u=rcos\theta, \,\,\, v = rsin\theta$

Interval changes as $u \in [0,\infty], \,\, v \in [0,\infty] \Longrightarrow r \in [0,\infty], \,\, \theta \in [0,\frac{\pi}{2}]$

Note: $dudv = rdrd\theta$ because $det(J_{(r,\theta)})=r$. Please see Jacobian.

$= 4\int^{\frac{\pi}{2}}_0\int^\infty_0 e^{-(r^2cos^2\theta+r^2sin^2\theta)} (rcos\theta)^{2a-1} (rsin\theta)^{2b-1} rdrd\theta$

$= 4 \int^\infty_0 e^{-r^2} r^{2a-1}r^{2b-1} rdr \cdot \int^{\frac{\pi}{2}}_0 cos^{2a-1}\theta sin^{2b-1}\theta d\theta$

$= 2 \int^\infty_0 r^{2(a+b)-1}e^{-r^2}dr \cdot 2 \int^{\frac{\pi}{2}}_0 cos^{2a-1}\theta sin^{2b-1}\theta d\theta$

$=\Gamma(a+b)B(b,a) = \Gamma(a+b)B(a,b)$

Hence,

$B(a,b) = \frac{\Gamma(a)\Gamma(b)}{\Gamma(a+b)}$

Note: Regarding $ \int A \int B = \int\int AB $ part, see Fubini's theorem.

## 4. Multivariate beta function

$B({\bf \alpha}) = \frac{\prod_{k=1}^K\Gamma(\alpha_k)}{\Gamma(\sum_{k=1}^K \alpha_k)}$

$where \,\,\, {\bf \alpha} = (\alpha_1,\alpha_2,...,\alpha_K)^T$
