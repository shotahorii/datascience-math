## 1. Binomial Theorem

$ (x+y)^N=\sum^N_{m=0} \left(\begin{array}{c}N\\m\\\end{array}\right)x^{N-m} y^m = \sum^N_{m=0} \frac{N!}{(N-m)!m!}x^{N-m} y^m$

$ where\,\,\,\, N,m \in \mathbb{N}_0 $

See [Binomial Coefficient](https://github.com/shotahorii/math-for-ds/blob/master/content/others/binomial_coefficient.ipynb)

#### Proof by induction

When $N=1$

$ x+y = \sum^1_{m=0} \frac{1!}{(1-m)!m!}x^{1-m} y^m = \frac{1!}{1!0!}x^1y^0 + \frac{1!}{0!1!}x^0 y^1 = x + y$

Proof that the theorem is true when $N=k+1$ if the theorem is true when $N=k$.

$(x+y)^k=\sum^k_{m=0} \left(\begin{array}{c}k\\m\\\end{array}\right)x^{k-m} y^m$

$\Longrightarrow (x+y)^{k+1}=(x+y) \sum^{k}_{m=0} \left(\begin{array}{c}k\\m\\\end{array}\right)x^{k-m} y^m$

$= \sum^{k}_{m=0} \left(\begin{array}{c}k\\m\\\end{array}\right)x^{k-m+1} y^m + \sum^{k}_{m=0} \left(\begin{array}{c}k\\m\\\end{array}\right)x^{k-m} y^{m+1}$

$= x^{k+1}+\sum^{k}_{m=1} \left(\begin{array}{c}k\\m\\\end{array}\right)x^{k-m+1} y^m + \sum^{k-1}_{m=0} \left(\begin{array}{c}k\\m\\\end{array}\right)x^{k-m} y^{m+1} + y^{k+1}$

Let: $m' = m+1$

$= x^{k+1}+\sum^{k}_{m=1} \left(\begin{array}{c}k\\m\\\end{array}\right)x^{k-m+1} y^m + \sum^{k}_{m'=1} \left(\begin{array}{c}k\\m'-1\\\end{array}\right)x^{k-m'+1} y^{m'} + y^{k+1}$

$= x^{k+1}+\sum^{k}_{m=1} \{ \left(\begin{array}{c}k\\m\\\end{array}\right) + \left(\begin{array}{c}k\\m-1\\\end{array}\right) \} x^{k-m+1} y^m + y^{k+1}$

$= x^{k+1}+\sum^{k}_{m=1} \left(\begin{array}{c}k+1\\m\\\end{array}\right) x^{k-m+1} y^m + y^{k+1}$

$= \sum^{k+1}_{m=0} \left(\begin{array}{c}k+1\\m\\\end{array}\right) x^{k+1-m} y^m$

## 2. Multinomial Theorem

The coefficient of $x_1^{m_1}x_2^{m_2}...x_K^{m_K}$ in the expansion of $(x_1 + x_2 + ... + x_K)^N$ is $\frac{N!}{m_1!m_2!...m_K!}$ 

Where $N, m_k \in \mathbb{N}_0 , \,\,\, \sum_{k=1}^K m_k = N$

In other words,

$(x_1 + x_2 + ... + x_K)^N = \sum_{\sum m_k = N} \frac{N!}{m_1!m_2!...m_K!} \prod_{k=1}^K x_k^{m_k}$

#### Proof by induction

When $K=1$

$x_1^N = \sum_{m_1 = N} \frac{N!}{m_1!} \prod_{k=1}^1 x_1^{m_1} = x_1^N$

Proof that the theorem is true when $K=r+1$ if the theorem is true when $K=r$.

$(x_1 + x_2 + ... + x_r + x_{r+1})^N = ((x_1 + x_2 + ... + x_r) + x_{r+1})^N$

Applying binomial theorem,

$= \sum_{i=0}^N \frac{N!}{i!(N-i)!} x_{r+1}^i(x_1 + x_2 + ... + x_r)^{N-i}$

Note: As assuming the theorem is true when $K=r$, $(x_1 + x_2 + ... + x_r)^{N-i} = \sum_{m_1+m_2+...+m_r = N-i} \frac{(N-i)!}{m_1!m_2!...m_r!} \prod_{k=1}^r x_k^{m_k}$

$= \sum_{i=0}^N \left( \frac{N!}{i!(N-i)!} x_{r+1}^i\sum_{m_1+m_2+...+m_r = N-i} \frac{(N-i)!}{m_1!m_2!...m_r!} \prod_{k=1}^r x_k^{m_k} \right)$

Renaming index variable $i$ to $m_{r+1}$

$= \sum_{m_{r+1}=0}^N \left( \frac{N!}{m_{r+1}!(N-m_{r+1})!} x_{r+1}^{m_{r+1}}\sum_{m_1+m_2+...+m_r = N-m_{r+1}} \frac{(N-m_{r+1})!}{m_1!m_2!...m_r!} \prod_{k=1}^r x_k^{m_k} \right)$

$= \sum_{m_{r+1}=0}^N \left(\sum_{m_1+m_2+...+m_r = N-m_{r+1}} \frac{N!}{m_{r+1}!(N-m_{r+1})!}\frac{(N-m_{r+1})!}{m_1!m_2!...m_r!} \prod_{k=1}^{r+1} x_k^{m_k} \right)$

$= \sum_{m_{r+1}=0}^N \left(\sum_{m_1+m_2+...+m_r = N-m_{r+1}} \frac{N!}{m_1!m_2!...m_r!m_{r+1}!} \prod_{k=1}^{r+1} x_k^{m_k} \right)$

Note: $\sum_{m_{r+1}=0}^N \left(\sum_{m_1+m_2+...+m_r = N-m_{r+1}} (\cdot) \right) = \sum_{m_1+m_2+...+m_r+m_{r+1} = N} (\cdot)$

$= \sum_{m_1+m_2+...+m_r+m_{r+1} = N} \frac{N!}{m_1!m_2!...m_r!m_{r+1}!} \prod_{k=1}^{r+1} x_k^{m_k}$
