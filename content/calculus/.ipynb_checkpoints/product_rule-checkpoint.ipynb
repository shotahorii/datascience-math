## 1. Product Rule

$(f\cdot g)' = f'\cdot g + f \cdot g'$

**Proof1**

Let: $h(x) = f(x)g(x)$

$h'(x) = \lim\limits_{\Delta x \to 0}\frac{h(x+\Delta x)-h(x)}{\Delta x}$

$= \lim\limits_{\Delta x \to 0}\frac{f(x+\Delta x)g(x+\Delta x)-f(x)g(x)}{\Delta x}$

$= \lim\limits_{\Delta x \to 0}\frac{f(x+\Delta x)g(x+\Delta x) - f(x)g(x+\Delta x) + f(x)g(x+\Delta x) -f(x)g(x)}{\Delta x}$

$= \lim\limits_{\Delta x \to 0}\frac{\{f(x+\Delta x)-f(x)\}g(x+\Delta x) + f(x)\{g(x+\Delta x) - g(x)\}}{\Delta x}$

$= \lim\limits_{\Delta x \to 0}\frac{f(x+\Delta x)-f(x)}{\Delta x} \lim\limits_{\Delta x \to 0}g(x+\Delta x)
+ \lim\limits_{\Delta x \to 0}f(x) \lim\limits_{\Delta x \to 0}\frac{g(x+\Delta x)-g(x)}{\Delta x}$

$=f'(x)g(x) + f(x)g'(x)$

Note: The fact that $\lim\limits_{\Delta x \to 0}g(x+\Delta x) = g(x)$ is deduced from a theorem that states that differentiable functions are continuous.

**Proof2**

Let: $h(u,v) = uv, \,\, u = f(x), \,\, v = g(x)$

With the chain rule,

$\frac{dh}{dx} = \frac{dh}{du}\frac{du}{dx} + \frac{dh}{dv}\frac{dv}{dx}$

$= v \cdot f'(x) + u \cdot g'(x)$

$= g(x)f'(x) + f(x)g'(x)$

## 2. General Leibniz rule

$(fg)^{(n)} = \sum_{k=0}^n \left(\begin{array}{c}n\\k\\\end{array}\right) f^{(n-k)}g^{(k)}$

**Proof**

The proof of the general Leibniz rule proceeds by induction. Let $f$ and $g$ be $n$-times differentiable functions.  
When $n=1$, the general Leibniz rule is true because of the usual product rule, as below.

$(fg)^{(1)} = \sum_{k=0}^1 \left(\begin{array}{c}1\\k\\\end{array}\right) f^{(1-k)}g^{(k)}
= \left(\begin{array}{c}1\\0\\\end{array}\right) f^{(1)}g^{(0)} + \left(\begin{array}{c}1\\1\\\end{array}\right) f^{(0)}g^{(1)} = f'g+fg'
$

Now, assume the general Leibniz rule is true when $n=N$, 

$(fg)^{(N)} = \sum_{k=0}^N \left(\begin{array}{c}N\\k\\\end{array}\right) f^{(N-k)}g^{(k)}$

Then,

$(fg)^{(N+1)} = \{\sum_{k=0}^N \left(\begin{array}{c}N\\k\\\end{array}\right) f^{(N-k)}g^{(k)}\}'$

Considering the linearity of differentiation,

$=\sum_{k=0}^N \left(\begin{array}{c}N\\k\\\end{array}\right) \{f^{(N-k)}g^{(k)}\}'$

$=\sum_{k=0}^N \left(\begin{array}{c}N\\k\\\end{array}\right) \{f^{(N-k+1)}g^{(k)} + f^{(N-k)}g^{(k+1)}\}$

$=\sum_{k=0}^N \left(\begin{array}{c}N\\k\\\end{array}\right) f^{(N-k+1)}g^{(k)} + \sum_{k=0}^N \left(\begin{array}{c}N\\k\\\end{array}\right) f^{(N-k)}g^{(k+1)}$

Shift the second term from $\sum_{k=0}^N$ to $\sum_{k=1}^{N+1}$

$=\sum_{k=0}^N \left(\begin{array}{c}N\\k\\\end{array}\right) f^{(N-k+1)}g^{(k)} + \sum_{k=1}^{N+1} \left(\begin{array}{c}N\\k-1\\\end{array}\right) f^{(N-k+1)}g^{(k)}$

$=\left(\begin{array}{c}N\\0\\\end{array}\right) f^{(N+1)}g^{(0)} + \sum_{k=1}^N \left(\begin{array}{c}N\\k\\\end{array}\right) f^{(N-k+1)}g^{(k)} + \sum_{k=1}^{N} \left(\begin{array}{c}N\\k-1\\\end{array}\right) f^{(N-k+1)}g^{(k)} + \left(\begin{array}{c}N\\N\\\end{array}\right) f^{(0)}g^{(N+1)}$

$=f^{(N+1)}g^{(0)} + \sum_{k=1}^N \{\left(\begin{array}{c}N\\k\\\end{array}\right)+\left(\begin{array}{c}N\\k-1\\\end{array}\right)\} f^{(N-k+1)}g^{(k)} + f^{(0)}g^{(N+1)}$

Using the [property of binomial coefficient](https://github.com/shotahorii/math-for-ds/blob/master/content/others/binomial_coefficient.ipynb),

$=f^{(N+1)}g^{(0)} + \sum_{k=1}^N \left(\begin{array}{c}N+1\\k\\\end{array}\right) f^{(N-k+1)}g^{(k)} + f^{(0)}g^{(N+1)}$

$=\sum_{k=0}^{N+1} \left(\begin{array}{c}N+1\\k\\\end{array}\right) f^{(N-k+1)}g^{(k)}$

Now proof is complete as the rule is true for $n=N+1$ when it's true for $n=N$

## 3. More than two factors

$$(f_1f_2...f_m)^{(n)} = \sum_{k_i \in \mathbb{N}_0, \sum_{i=1}^m k_i = n} \left(\frac{n!}{\prod_{i=1}^mk_i!}\prod_{i=1}^mf_i^{(k_i)}\right)$$

where the sum extends over all m-tuples $(k_1,...,k_m)$ of non-negative integers with $\sum_{i=1}^mk_i! = n$

**Example**

Let: $n=1$, $m=3$ then,

$(f_1f_2f_3)' = \sum_{k_i \in \mathbb{N}_0, \sum_{i=1}^3 k_i = 1} \left(\frac{1!}{\prod_{i=1}^3k_i!}\prod_{i=1}^3f_i^{(k_i)}\right)$

Note: the sum here extends over $(k_1=1,k_2=0,k_3=0)$, $(k_1=0,k_2=1,k_3=0)$ and $(k_1=0,k_2=0,k_3=1)$

$= \left(\frac{1!}{1!0!0!}f_1^{(1)}f_2^{(0)}f_3^{(0)} \right) + \left(\frac{1!}{0!1!0!}f_1^{(0)}f_2^{(1)}f_3^{(0)} \right) + \left(\frac{1!}{0!0!1!}f_1^{(0)}f_2^{(0)}f_3^{(1)} \right)$

$= f_1'f_2f_3 + f_1f_2'f_3 + f_1f_2f_3'$
