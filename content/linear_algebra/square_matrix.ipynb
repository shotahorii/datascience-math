## 1. Square Matrix

A square matrix is a matrix with the same number of rows and columns. An $n$-by-$n$ matrix is known as a square matrix of order $n$.

### Operations defined for square matrix
- Inverse
- Determinant
- Trace
- Eigenvalues and eigenvectors

### Types of matrix which are square matrix by definition
- Diagonal Matrix
- Identity Matrix
- Regular Matrix
- Symmetric Matrix
- Triangular matrix
- Orthogonal matrix
- Idempotent matrix

## 2. Diagonal Matrix

A diagonal matrix $D$ is a square matrix in which the entries outside the main diagonal are all zero. By definition, a diagonal matrix is a symmetric matrix.

$D= diag(d_1,d_2,\cdots,d_n) = 
\begin{bmatrix}
d_1 & 0 & \cdots & 0 \\
0 & d_2 & \cdots & 0 \\
\vdots & & \ddots & \vdots \\
0 & 0 & \cdots & d_n \\
\end{bmatrix}$

### Properties 

**1.** 

Let: $A= diag(a_1,a_2,...,a_n), B= diag(b_1,b_2,...,b_n)$

Then: $AB=BA=diag(a_1b_1,a_2b_2,...,a_nb_n)$

**2.** 

$D^{m} = diag(d_1,d_2,...,d_n)^{m} = diag(d_1^{m},d_2^{m},...,d_n^{m})$

**3.** 

$D^T = D$

This means that a diagonal matrix is also a symmetric matrix.

**4.** 

$D^{-1} = diag(d_1,d_2,...,d_n)^{-1} = diag(d_1^{-1},d_2^{-1},...,d_n^{-1})$

**5.**

$det(D) = \prod_{i=1}^nd_i$

**6.**

$D$ is a regular matrix $\Longleftrightarrow$ all of $d_i$ are non-zero

(This is obvious from property 5. above and definition of regular matrix.)

**7.**

$d_1,d_2,...,d_n$ are the eigenvalues of $D = diag(d_1,d_2,...,d_n)$

Proof below:

$\lambda I - D = 
\begin{bmatrix}
\lambda_1 & 0 & \cdots & 0 \\
0 & \lambda_2 & \cdots & 0 \\
\vdots & & \ddots & \vdots\\
0 & 0 & \cdots & \lambda_n \\
\end{bmatrix}
- 
\begin{bmatrix}
d_1 & 0 & \cdots & 0 \\
0 & d_2 & \cdots & 0 \\
\vdots & & \ddots &\vdots \\
0 & 0 & \cdots & d_n \\
\end{bmatrix}
=
\begin{bmatrix}
\lambda_1 - d_1 & 0 & \cdots & 0 \\
0 & \lambda_2 - d_2 & \cdots & 0 \\
\vdots & & \ddots &\vdots \\
0 & 0 & \cdots & \lambda_n - d_n \\
\end{bmatrix}
$

Hence, $|\lambda I - D| = (\lambda_1 - d_1)(\lambda_2 - d_2)...(\lambda_n - d_n)$

Hence $\lambda_i = d_i$ , considering the characteristic equation $|\lambda I - D|=0$

## 3. Identity Matrix

An identity matrix $I$ is a diagonal matrix in which all diagonal elements are 1.

$I=
\begin{bmatrix}
1 & 0 & \cdots & 0 \\
0 & 1 & \cdots & 0 \\
\vdots & & \ddots & \vdots\\
0 & 0 & \cdots & 1 \\
\end{bmatrix}$

## 4. Regular Matrix

### Definition

A $n \times n$ square matrix $A$ is a regular matrix if any of below is true. (5 conditions below are all equal. So if one is true, then all are ture.)

**(1)** There's a square matrix $B$ such that $AB=BA=I$. Then, $B$ is called an inverse matrix of $A$ and expressed as $A^{-1}$.

**(2)** $det(A) \ne 0$

**(3)** $rank(A) = n$

**(4)** $ker(A) = \{{\bf 0}\}$

**(5)** All eigenvalues of $A$ are non-zero

### Properties

**1.**

When $A$ is a regular matrix, $A^{-1}$ is also a regular matrix and below is true.

$(A^{-1})^{-1} = A$

**2.**

When $A$ and $B$ are $n \times n$ regular matrix, $AB$ is also a regular matrix. And below is true.

$(AB)^{-1} = B^{-1}A^{-1}$

**3.**

When $A$ is a regular matrix, $A^T$ is also a regular matrix, and below is true.

$(A^T)^{-1} = (A^{-1})^T$

## 5. Inverse Matrix

$A^{-1} = \frac{1}{|A|}\tilde{A}$

where $\tilde{A}$ is the cofactor matrix of $A$.

**Example**

Let: $A = \begin{bmatrix}
a & b \\
c & d \\
\end{bmatrix}$

$A^{-1} = \frac{1}{ad-bc}
\begin{bmatrix}
d & -b \\
-c & a \\
\end{bmatrix}$
