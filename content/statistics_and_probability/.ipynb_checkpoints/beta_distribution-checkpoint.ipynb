## 1. Definition

$ Beta(\mu|a,b)= \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\mu^{a-1}(1-\mu)^{b-1} $

$ where\,\,\,\, a,b \in \mathbb{R}^+ \,\,\,\, \mu \in [0,1] $

---

## 2. Proof that Binomial distribution is normalised

$ \int^1_0 Beta(\mu|a,b) d\mu = \int^1_0 \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\mu^{a-1}(1-\mu)^{b-1} d\mu $

$ = \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)} \int^1_0 \mu^{a-1}(1-\mu)^{b-1} d\mu $

$ = \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)} B(a,b)$

$ = \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)} \frac{\Gamma(a)\Gamma(b)}{\Gamma(a+b)} = 1$

See [Gamma function and beta function](https://github.com/shotahorii/math-for-ml/blob/master/content/fundamental/gamma_function_and_beta_function.ipynb)

---

## 3. Expected value

$E[\mu] =\int^1_0 \mu Beta(\mu|a,b) d\mu = \int^1_0 \mu \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\mu^{a-1}(1-\mu)^{b-1} d\mu$

$= \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\int^1_0 \mu^{(a+1)-1}(1-\mu)^{b-1} d\mu$

$= \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}B(a+1,b) 
= \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)} \frac{\Gamma(a+1)\Gamma(b)}{\Gamma(a+b+1)}$

$= \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)} \frac{a \Gamma(a)\Gamma(b)}{(a+b) \Gamma(a+b)} = \frac{a}{a+b}$

---

## 4. Variance

$E[\mu^2] = \int^1_0 \mu^2 Beta(\mu|a,b) d\mu = \int^1_0 \mu^2 \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\mu^{a-1}(1-\mu)^{b-1} d\mu$

$= \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}\int^1_0 \mu^{(a+2)-1}(1-\mu)^{b-1} d\mu$

$= \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)}B(a+2,b) 
= \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)} \frac{\Gamma(a+2)\Gamma(b)}{\Gamma(a+b+2)}$

$= \frac{\Gamma(a+b)}{\Gamma(a)\Gamma(b)} \frac{a(a+1)\Gamma(a)\Gamma(b)}{(a+b+1)(a+b)\Gamma(a+b)} $

$= \frac{a(a+1)}{(a+b+1)(a+b)}$

Hence,

$Var[\mu] = E[\mu^2] - (E[\mu])^2$

$= \frac{a(a+1)}{(a+b+1)(a+b)} - (\frac{a}{a+b})^2 $

$= \frac{ab}{(a+b)^2(a+b+1)}$
