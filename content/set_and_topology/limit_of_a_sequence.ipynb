## 1.Intuitive definition and examples
When there's a number corresponding to each of natural numbers $1,2,3,...,n,...$, the ordered list of the corresponding numbers $a_1,a_2,a_3,...,a_n,...$ is called a sequence. And it's represented as $\{a_n\}$.

Having a sequence $\{a_n\}$, when $a_n$ gets arbitrarily close to a specific value $\alpha$ (=the limit) as $n$ approaches infinity, we say that $\{a_n\}$ converges to $\alpha$, and the sequence $\alpha$ is called **convergent**. This is expressed as below.

$\lim\limits_{n \to \infty}a_n = \alpha$

Below are some examples of convergent.

**Example 1**: A sequence $\{\frac{1}{n}\}$, which is $1,\frac{1}{2},\frac{1}{3},...,\frac{1}{n},...$

$\lim\limits_{n \to \infty}\frac{1}{n} = 0$

**Example 2**: A sequence $\{\frac{n^2+3}{3n^2+4n-1}\}$, which is $\frac{2}{3},\frac{7}{19},\frac{6}{19},...,\frac{n^2+3}{3n^2+4n-1},...$

$\lim\limits_{n \to \infty}\frac{n^2+3}{3n^2+4n-1} 
= \lim\limits_{n \to \infty}\frac{1+\frac{3}{n^2}}{3+\frac{4}{n}-\frac{1}{n^2}} 
= \frac{1+0}{3+0-0} = \frac{1}{3}$


When the sequence doesn't have a limit, it's called **divergent**. Below are some examples of divergent.

**Example 3**: A sequence $\{n\}$, which is $1,2,3,...,n,...$

$\lim\limits_{n \to \infty}n = \infty$

**Example 4**: A sequence $\{-n\}$, which is $-1,-2,-3,...,-n,...$

$\lim\limits_{n \to \infty}-n = -\infty$

**Example 5**: A sequence $\{n(-1)^n\}$, which is $-1,2,-3,...,n(-1)^n,...$

This goes ${\to -\infty}$ when $n$ is an odd number, and ${\to \infty}$ when $n$ is an even number.

**Example 6**: A sequence $\{(-1)^n\}$, which is $-1,1,-1,...,(-1)^n,...$

This goes ${\to -1}$ when $n$ is an odd number, and ${\to 1}$ when $n$ is an even number.

## 2. Precise definition

$\lim\limits_{n \to \infty}a_n = \alpha \Longleftrightarrow 
(\forall \varepsilon >0,\exists N \in \mathbb{N} \, s.t. \, \forall n \in \mathbb{N}, n \ge N \Rightarrow |a_n - \alpha| < \varepsilon)$


**Example 7**: Proof of $\lim\limits_{n \to \infty}(2-\frac{1}{n})=2$

To proof this, we need to show that for any $\varepsilon > 0$, there's $N \in \mathbb{N}$ such that $|2-\frac{1}{n}-2| < \varepsilon $ when $n \ge N$.  
Now, choose a natural number $M > \frac{1}{\varepsilon}$. Note that by pre-condition, $n \ge M$ hence $\frac{1}{n} \le \frac{1}{M}$.

Then, $|2-\frac{1}{n}-2|=|-\frac{1}{n}|=\frac{1}{n} \le \frac{1}{M} < \varepsilon$

## 3. Some theorems 
**Theorem 1**: A partial sequence of a convergent converges to the convergent's limit.  
Note that, a partial sequence of a divergent can be a convergent. For example, A sequence $\{(-1)^n\}$ is a divergent, but a partial sequence $a_2, a_4, ...$ converges to $1$ and a partial sequence $a_1, a_3, ...$ converges to $-1$.

**Theorem 2**: $\lim\limits_{n \to \infty}a_n = \alpha \Longrightarrow (\exists M \in \mathbb{R}^+ \,s.t.\, \forall n \in \mathbb{N},|a_n|<M, |\alpha|<M)$  
This means: If a sequence is a convergent, it's bounded and its limit doesn't go beyond the upper/lower bound.  
However, note that a bounded sequence is not always a convergent. For example $\{(-1)^n\}$.

A sequence $\{a_n\}$ is bounded $\Longleftrightarrow (\exists L \in \mathbb{R}, \exists U \in \mathbb{R} \, s.t. \, \forall n \in \mathbb{N}, L \le a_n \le U )$

**Theorem 3**: When $\{a_n\}$ and $\{b_n\}$ are convergents, below formulas are true.   

(1) $\lim\limits_{n \to \infty}(a_n+b_n) = \lim\limits_{n \to \infty}a_n + \lim\limits_{n \to \infty}b_n$

(2) $\lim\limits_{n \to \infty}(a_n-b_n) = \lim\limits_{n \to \infty}a_n - \lim\limits_{n \to \infty}b_n$

(3) $\lim\limits_{n \to \infty}(a_nb_n) = (\lim\limits_{n \to \infty}a_n)(\lim\limits_{n \to \infty}b_n)$

(4) $\lim\limits_{n \to \infty}\frac{a_n}{b_n} = \frac{\lim\limits_{n \to \infty}a_n}{\lim\limits_{n \to \infty}b_n}
\,\,\, where \,\,\, b_n \ne 0, \lim\limits_{n \to \infty}b_n \ne 0$

**Theorem 4**: if a sequence $\{a_n\}$ is monotone increasing/decreasing AND bounded, it's a convergent.
