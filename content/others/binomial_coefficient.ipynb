## 1. Definition
- The coefficient of $x^m$ in the expansion of $(1 + x)^N$
- The number of ways, disregarding order, that $m$ objects can be chosen from among $N$ objects
- The formula is below.

$\left(\begin{array}{c}N\\m\\\end{array}\right)
= \frac{N!}{(N-m)!m!}
$

$where \,\,\,\, N,m \in \mathbb{N}_0 \,\,\,\, m \le N$

Also, by the definition, we have below formula. 

$(1+x)^N = \sum^N_{m=0}\left(\begin{array}{c}N\\m\\\end{array}\right)x^m$

## 2. Frequently Used Equations 
### Equation 1

$\left(\begin{array}{c}N\\m\\\end{array}\right) 
= \left(\begin{array}{c}N\\N-m\\\end{array}\right)$

#### Proof

$\left(\begin{array}{c}N\\m\\\end{array}\right) = \frac{N!}{(N-m)!m!} 
= \frac{N!}{(N-m)!\{N-(N-m)\}!} = \left(\begin{array}{c}N\\N-m\\\end{array}\right)$

---

### Equation 2

$\left(\begin{array}{c}N\\m\\\end{array}\right)
=\left(\begin{array}{c}N-1\\m\\\end{array}\right)
+\left(\begin{array}{c}N-1\\m-1\\\end{array}\right)$

#### Proof

$\left(\begin{array}{c}N-1\\m\\\end{array}\right)
+\left(\begin{array}{c}N-1\\m-1\\\end{array}\right)
= \frac{(N-1)!}{(N-1-m)!m!} + \frac{(N-1)!}{\{N-1-(m-1)\}(m-1)!}$

$=\frac{(N-1)!}{(N-1-m)!(m-1)!}(\frac{1}{m}+\frac{1}{N-m})$

$=\frac{(N-1)!}{(N-m-1)!(m-1)!}\frac{N}{(N-m)m}$

$=\frac{N!}{(N-m)!m!}=
\left(\begin{array}{c}N\\m\\\end{array}\right)$

---

### Equation 3
$\left(\begin{array}{c}N\\m\\\end{array}\right)
= \frac{N}{m}\left(\begin{array}{c}N-1\\m-1\\\end{array}\right)$

#### Proof

$\left(\begin{array}{c}N\\m\\\end{array}\right)=\frac{N!}{(N-m)!m!}$

$=\frac{N}{m}\frac{(N-1)!}{(N-m)!(m-1)!}$

$=\frac{N}{m}\frac{(N-1)!}{\{(N-1)-(m-1)\}!(m-1)!}$

$=\frac{N}{m}\left(\begin{array}{c}N-1\\m-1\\\end{array}\right)$
